## Wikidata SPARQL

Want to extract wikipedia page from id after query (e.g., 'Q***')

```python
endpoint_url = "https://query.wikidata.org/sparql"

query = """SELECT ?child ?childLabel
WHERE
{
  ?child wdt:P22 wd:Q1339;
         wdt:P25 wd:Q57487.
  SERVICE wikibase:label { bd:serviceParam wikibase:language "en". }
}"""

def get_results(endpoint_url, query):
    user_agent = "WDQS-example Python/%s.%s" % (sys.version_info[0], sys.version_info[1])
    # TODO adjust user agent; see https://w.wiki/CX6
    sparql = SPARQLWrapper(endpoint_url, agent=user_agent)
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)
    return sparql.query().convert()

results = get_results(endpoint_url, query)

for result in results["results"]["bindings"]:
    print(result['child'])
    print(result['childLabel'])
```

# Wikipedia api to extract page, t-SNE

- https://towardsdatascience.com/google-news-and-leo-tolstoy-visualizing-word2vec-word-embeddings-with-t-sne-11558d8bd4d

```python
title = 'String theory'
response = requests.get(
    'https://en.wikipedia.org/w/api.php',
    params={
        'action': 'query',
        'format': 'json',
        'titles': title,
        'prop': 'extracts',
        'explaintext': True,
    } ).json()
page = next(iter(response['query']['pages'].values()))
try:
    end_ = page['extract'].index('== See also ==')
except:
    end_ = page['extract'].index('== Notes ==')
    pass
text = page['extract'][:end_]

# with open("wiki_page.txt", "w") as f:
#     f.write(text)

all_sentences = nltk.sent_tokenize(text)
all_words = [nltk.word_tokenize(sent) for sent in all_sentences]

for i in range(len(all_words)):
    all_words[i] = [w for w in all_words[i] if w not in stopwords.words('english')]
    
model = Word2Vec(all_words, min_count=10, size=50, window=10)

keys = ['theory', 'geometry', 'physics', 'symmetry']
embedding_clusters = []
word_clusters = []
for word in keys:
    embeddings = []
    words = []
    for similar_word, _ in model.most_similar(word, topn=30):
        words.append(similar_word)
        embeddings.append(model[similar_word])
    embedding_clusters.append(embeddings)
    word_clusters.append(words)
    
embedding_clusters = np.array(embedding_clusters)
```

```python
n, m, k = embedding_clusters.shape
tsne_model_en_2d = TSNE(perplexity=30, n_components=2, init='pca', n_iter=3500)
embeddings_en_2d = np.array(tsne_model_en_2d.fit_transform(embedding_clusters.reshape(n * m, k))).reshape(n, m, 2)
```

```python
def tsne_plot_similar_words(labels, embedding_clusters, word_clusters, name):
    plt.figure(figsize=(16, 9))
    #colors = cm.rainbow(np.linspace(0, 1, len(labels)))
    #for label, embeddings, words, color in zip(labels, embedding_clusters, word_clusters, colors):
    for label, embeddings, words in zip(labels, embedding_clusters, word_clusters):
        x = embeddings[:, 0]
        y = embeddings[:, 1]
        plt.scatter(x, y, alpha=0.7, label=label)
        for i, word in enumerate(words):
            plt.annotate(word, alpha=0.5, xy=(x[i], y[i]), xytext=(5, 2),
                         textcoords='offset points', ha='right', va='bottom', size=8)
    plt.legend(loc=4)
    if name:
        plt.savefig(f'sne {name}')

tsne_plot_similar_words(keys, embeddings_en_2d, word_clusters,i)
```
